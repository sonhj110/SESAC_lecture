```python
import pandas as pd

topictable = pd.read_csv('/content/drive/MyDrive/SK하이닉스/10. 기사 추천시스템/topictable_notfiltered.csv', header = 0)
```

```python
topictable.drop(['Unnamed: 0', '각 토픽의 비중'], axis=1, inplace=True)
topictable['가장 비중이 높은 토픽'] = topictable['가장 비중이 높은 토픽'] + 1
topictable.drop(['가장 비중이 높은 토픽'], axis=1, inplace=True)
topictable
```

```python
news = pd.read_csv('/content/drive/MyDrive/SK하이닉스/07. LASSO 회귀/filtered_sk_data.csv', header = 0)
```

```python
news.rename(columns={'Unnamed: 0': '문서 번호'}, inplace=True)
news
```

```python
merged_df = pd.merge(news, topictable, on='문서 번호', how='left')
merged_df = merged_df.drop(['tokens'], axis=1)
merged_df
```

```python
merged_df['date'] = pd.to_datetime(merged_df['date'])
merged_df.info()
```

```python
peak = pd.read_csv('/content/drive/MyDrive/SK하이닉스/04. 피크데이 계산/PeakDays/PeakDays.csv')
peak
```

```python
peak['date'] = pd.to_datetime(peak['Date'], format='%Y%m%d')
peak.info()
```

```python
peak.drop(['Date'],  axis=1, inplace=True)
```

```python
peakday_list = peak['date']
peakday_list
```

```python
mask = merged_df['date'].isin(peakday_list)
peak_filtered_df = merged_df[mask]
peak_filtered_df
```

```python
peak_filtered_df.to_csv('/content/drive/MyDrive/SK하이닉스/10. 기사 추천시스템/peak_rec_articles.csv', index=False)
```

```python
# 피크일에 올라온 해당 토픽의 뉴스 보여주기

topic_list = [1,2,3,4,6,7,9,10,11,14,15,16,19,20,21,28,33,35,39,44,47,49]
# peakday_list
topicNo = 47

topic = peak_filtered_df[peak_filtered_df['가장 비중이 높은 토픽'] == topicNo]
peak_topic_news = topic.sort_values(by=['date', '가장 높은 토픽의 비중'], ascending=[True, False])
peak_topic_news[peak_topic_news['date'] == peakday_list[3]]
```

```python
# 피크일 전체에 올라온 해당 토픽의 뉴스 개수

topic_list = [1,2,3,4,6,7,9,10,11,14,15,16,19,20,21,28,33,35,39,44,47,49]
# topicNo = 44
for topicNo in topic_list :
  topic = peak_filtered_df[peak_filtered_df['가장 비중이 높은 토픽'] == topicNo]
  peak_topic_news = topic.sort_values(by=['date', '가장 높은 토픽의 비중'], ascending=[True, False])
  print(topicNo, len(peak_topic_news))
```

```python
peak_topic = pd.read_csv('/content/drive/MyDrive/SK하이닉스/10. 기사 추천시스템/peakdays_dominant_topic.csv')
peak_topic['date'] = pd.to_datetime(peak_topic['date'])
peak_topic.info()
```

```python
def onlynum(mystr) :
  mystr = mystr.replace('topic', "")
  return(int(mystr))

peak_topic['피크일을 가장 잘 예측한 토픽'] = peak_topic['피크일을 가장 잘 예측한 토픽'].apply(onlynum)
peak_topic.info()
```
